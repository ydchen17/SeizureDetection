```python
# twint: https://github.com/twintproject/twint
# nest_asyncio: https://stackoverflow.com/questions/67585628/runtimeerror-in-twint-this-event-loop-is-already-running

# twint is not being well maintained for a while, so there is issues with its PyPI source, where as to solve this issue,
# the only way we can do is to install from GitHub repository. Read here for the issue:
# https://github.com/twintproject/twint/issues/915#issuecomment-716644446
# also read: https://github.com/twintproject/twint/issues/1266

import twint
import nest_asyncio
import twython
import numpy as np
import re
import emoji
import datetime

nest_asyncio.apply()
# Search stock
subject = "$AAPL"
# Configuring Twint to search for the subject in the first specified city
c = twint.Config()
# The amount of tweets to return sorted by most recent
c.Limit = 100
# Input parameters
c.Search = subject
# Removing retweets
c.Filter_retweets = True
# No pictures or video
c.Media = False
# English only
c.Lang ='en'
# Excluding tweets with links
c.limit = 200
c.Links ='exclude'
c.Since = '2020-10-01' #"2020-10-01 00:00:00"
c.Until = '2020-10-02' #"2020-11-01 00:00:00"
# Making the results pandas friendly
c.Pandas = True
twint.run.Search(c)
# Assigning the DF
df = twint.storage.panda.Tweets_df
print(df.shape)
```
